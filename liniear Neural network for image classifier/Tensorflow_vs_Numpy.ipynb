```python
import numpy as np
import tensorflow as tf
import time
```

```python
n = 10000 # increase this as you need to see the performance tf vs np
```

```python
A_np = np.random.rand(n,n)
B_np = np.random.rand(n,n)
```

```python
#convert numpy matrices to tensorflow tensors
A_tf = tf.convert_to_tensor(A_np, dtype=tf.float32)
B_tf = tf.convert_to_tensor(B_np, dtype=tf.float32)
```

```python
# Numpy matrix calculation on CPU
start_time = time.time()
C_np = np.dot(A_np, B_np)
end_time = time.time()
np_time = abs(start_time - end_time)
print(f"numpy time is {np_time:.6f} seconds")
```

```python
# TensorFlow matrix calculation on GPU
start_time = time.time()
with tf.device('/GPU:0'): # to get the usage of the GPU
  C_tf = tf.matmul(A_tf, B_tf)
end_time = time.time()
tensorflow_time = end_time - start_time
print(f"TensorFlow time is on GPU :{tensorflow_time:.6f} seconds")
```
